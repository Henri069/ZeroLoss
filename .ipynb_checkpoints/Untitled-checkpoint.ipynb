```python
import pandas as pd
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
from sklearn.model_selection import train_test_split

```

```python
#Data Cleaning 
#Load data

df = pd.read_csv('lendingclubLoanData.csv')

#Select Features 
df1 = df[['loan_amnt','term','int_rate','installment','grade','emp_length','home_ownership','annual_inc','verification_status','zip_code','loan_status']]
#Filter Loan status and drop all rows with null loan status 
df1 = df1[(df1['loan_status'] != 'Current') & (df1['loan_status'] != 'In Grace Period') 
          & (df1['loan_status'] != 'Late (31-120 days)') & (df1['loan_status'] != 'Late (16-30 days)')& (df1['loan_status'].notna())]
#Drop all empty rows 
df1.dropna(how = 'all', inplace = True)
#Fill all NaN with 0
df1.fillna(0,inplace = True)

#Change string features into int 
df1.loc[df1['grade'] == 'G', 'grade'] = 1
df1.loc[df1['grade'] == 'F', 'grade'] = 2
df1.loc[df1['grade'] == 'E', 'grade'] = 3
df1.loc[df1['grade'] == 'D', 'grade'] = 4
df1.loc[df1['grade'] == 'C', 'grade'] = 5
df1.loc[df1['grade'] == 'B', 'grade'] = 6
df1.loc[df1['grade'] == 'A', 'grade'] = 7

df1.loc[df1['home_ownership'] == 'ANY', 'home_ownership'] = 1
df1.loc[df1['home_ownership'] == 'MORTGAGE', 'home_ownership'] = 2
df1.loc[df1['home_ownership'] == 'NONE', 'home_ownership'] = 3
df1.loc[df1['home_ownership'] == 'OWN', 'home_ownership'] = 4
df1.loc[df1['home_ownership'] == 'RENT', 'home_ownership'] = 5

df1.loc[df1['verification_status'] == 'Not Verified', 'verification_status'] = 1
df1.loc[df1['verification_status'] == 'Verified', 'verification_status'] = 2
df1.loc[df1['verification_status'] == 'Source Verified', 'verification_status'] = 3

df1.loc[df1['loan_status'] == 'Fully Paid', 'loan_status'] = 0
df1.loc[df1['loan_status'] == 'Default', 'loan_status'] = 1
df1.loc[df1['loan_status'] == 'Charged Off', 'loan_status'] = 1

#Train test split
train, test = train_test_split(df1, test_size=0.2, shuffle=True)

#Save to CSV
train.to_csv('train.csv',index = False, header = False)
test.to_csv('test.csv',index = False, header = False)


# load data
dataset = loadtxt('train.csv', delimiter=',')
# split features and target
X = dataset[:,0:10]
y = dataset[:,10]

# Create Keras Model

model = Sequential()
model.add(Dense(320, input_dim=10, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# compile
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# fit model
model.fit(X, y, epochs=150, batch_size=10)
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```
